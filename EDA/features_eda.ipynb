```python
import pandas as pd 
import talib as ta
import matplotlib.pyplot as plt
from matplotlib import rcParams
import os

rcParams['axes.titlepad'] = 20
# No warnings about setting value on copy of slice
pd.options.mode.chained_assignment = None
```

```python
#create new folder to save results of EDA
current_directory = os.getcwd()
final_directory = os.path.join(current_directory, 'OHLCV_EDA_plots')
final_directory
if not os.path.exists(final_directory):
    os.makedirs(final_directory)
```

```python
df = pd.read_csv("../Data-Acquisition/ohlcv-data/ohlcv.csv")
```

```python
df_copy = df.copy()
df_copy["time"] = pd.to_datetime(df_copy["time"], format='%d/%m/%Y') 
```

```python
df_comp = df_copy[df_copy["symbol"] == "AMZN"]
#Momemtum indicator - 1
df_comp["RSI"] = ta.RSI(df_comp.close, timeperiod = 90)
#Momemtum indicator - 2
df_comp["ROC"] = ta.ROC(df_comp.close, timeperiod = 90)
#volume indicator
df_comp["ADOSC"] = ta.ADOSC(df_comp.high, df_comp.low, df_comp.close, df_comp.volume, fastperiod=30, slowperiod=90)
#volatility indicator
df_comp["ATR"] = ta.ATR(df_comp.high, df_comp.low, df_comp.close, timeperiod=90)
#trend indicator
df_comp["EMA"] = ta.EMA(df_comp.close, timeperiod = 90)
df_comp["time"] = pd.to_datetime(df_comp.time)
```

```python
plt.figure(figsize=(25, 10))
plt.plot(df_comp["time"][90:].tolist(), df_comp["open"][90:].tolist(), color="magenta")
plt.plot(df_comp["time"][90:].tolist(), df_comp["close"][90:].tolist(), color="green")
plt.legend(["open", "close"], bbox_to_anchor=(1.05, 1), loc=1, borderaxespad=0.)
plt.title("open and close of Amazon Stock in 2 year period", fontsize=20)
str_title = "open and close of Amazon Stock in 2 year period"
plt.savefig(final_directory+"/"+str_title, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.plot(df_comp["time"][90:].tolist(), df_comp["EMA"][90:].tolist(), color="red")
plt.legend(["EMA"], bbox_to_anchor=(1.05, 1), loc=1, borderaxespad=0.)
plt.title("Plot of EMA on Amazon Stock in 2 year period", fontsize=20)
str_title = "Plot of EMA on Amazon Stock in 2 year period"
plt.savefig(final_directory+"/"+str_title, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.plot(df_comp["time"][90:].tolist(), df_comp["ATR"][90:].tolist(), color="springgreen")
plt.legend(["ATR"], bbox_to_anchor=(1.05, 1), loc=1, borderaxespad=0.)
plt.title("Plot of ATR on Amazon Stock in 2 year period", fontsize=20)
str_title = "Plot of ATR on Amazon Stock in 2 year period"
plt.savefig(final_directory+"/"+str_title, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.plot(df_comp["time"][90:].tolist(), df_comp["ADOSC"][90:].tolist(), color="magenta")
plt.legend(["ADOSC"], bbox_to_anchor=(1.05, 1), loc=1, borderaxespad=0.)
plt.title("Plot of ADOSC on Amazon Stock in 2 year period", fontsize=20)
str_title = "Plot of ADOSC on Amazon Stock in 2 year period"
plt.savefig(final_directory+"/"+str_title, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.plot(df_comp["time"][90:].tolist(), df_comp["close"][90:].tolist(), color="green")
plt.legend(["ROC"], bbox_to_anchor=(1.05, 1), loc=1, borderaxespad=0.)
plt.title("Plot of ROC on Amazon Stock in 2 year period", fontsize=20)
str_title = "Plot of ROC on Amazon Stock in 2 year period"
plt.savefig(final_directory+"/"+str_title, bbox_inches="tight")
plt.show()
```
